```python
%matplotlib inline

import json
import numpy as np
import matplotlib.pyplot as plt
from os.path import isfile, join, isdir
from shapely.geometry import asShape
from descartes import PolygonPatch
from scipy.ndimage import imread
from scipy.misc import imresize
from os import listdir
```

### Average map

```python
models_paths = [f for f in listdir('./data/flags/colombia/') if isfile(join('./data/flags/colombia/', f))]
models_paths = sorted(models_paths)
```

```python
flags = list(map(lambda x:x.replace('.png', ''), models_paths))
```

```python
flags_dict = {}
```

```python
for f in flags:
    image = imread('./data/flags/colombia/%s.png' % f, mode='RGB')
    image = image.reshape((image.shape[0]*image.shape[1], 3))
    flags_dict[f] = np.mean(image, axis=-2)/255
```

```python
data = json.load(open("./data/colombia_std_names.geojson"))
```

```python
fig = plt.figure(figsize=(9, 10))
plt.axis('off')
ax = fig.gca() 

for feat in data["features"]:
    geom = asShape(feat["geometry"])
    x, y = geom.centroid.x, geom.centroid.y
    ax.plot(x, y, '-')
    ax.add_patch(PolygonPatch(feat["geometry"], fc=flags_dict[feat['properties']['STD_ID']], ec='black',
                alpha=1, lw=1, ls='-', zorder=2))
ax.clear
```

### Average flag

```python
complete_flags_array = []
for i, f in enumerate(flags):
    image = imread('./data/flags/colombia/%s.png' % f, mode='RGB')
    complete_flags_array.append(imresize(image,(200,200)))
complete_flags_array = np.array(complete_flags_array)
```

```python
complete_flags_array.shape
```

```python
mean_flag = np.mean(complete_flags_array, axis=-4).astype(int)
plt.axis('off')
plt.imshow(mean_flag)
```

```python
median_flag = np.median(complete_flags_array, axis=-4).astype(int)
plt.axis('off')
plt.imshow(median_flag)
```
