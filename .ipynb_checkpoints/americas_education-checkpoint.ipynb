```python
def fix(x, out):
    #print(x)
    year = 2016
    cur_val = None
    while year > 2000:
        cur_val = x[str(year)]
        if cur_val == cur_val:
            if out == "year":
                return year
            else:
                return cur_val
        else:
            year = year - 1
    if out == "year":
        return year
    else:
        return cur_val
```

```python
#https://data.worldbank.org/indicator/SE.XPD.TOTL.GD.ZS
df = pd.read_csv("./data/gasto_ed.csv")
df["year"] = df.apply(lambda x: fix(x, "year"), axis=1)
df["value"] = df.apply(lambda x: fix(x, "value"), axis=1)
df = df[["Country Name","Country Code","year","value"]]
df.index = df["Country Code"]
```

```python
spaces = 5
max_value = df["value"].max()
print(max_value)
unknown_value = (128,128,128)
bottom_value = (0,0,100)
top_value = (255,255,255)
arr = [bottom_value]*100*100
diff = {}
diff['r'] = abs(top_value[0] - bottom_value[0])//spaces
diff['g'] = abs(top_value[1] - bottom_value[1])//spaces
diff['b'] = abs(top_value[2] - bottom_value[2])//spaces
colors = []
colors.append(bottom_value)
for i in range(spaces - 1):
    color = ((bottom_value[0]+diff['r']*(i+1)),(bottom_value[1]+diff['g']*(i+1)),(bottom_value[2]+diff['b']*(i+1)))
    arr = arr + [color]*100*100
    colors.append(color)
arr = np.array(arr).reshape(spaces*100,100,3)
ranges_dict = {}
for i in range(spaces):
    ranges_dict[(round(max_value/spaces*i,2), round(max_value/spaces*(i+1),2))] = colors[-1-i]
plt.imshow(arr)
```

```python
def polygon_from_centroid(cen, u=100):
    return [
        [cen[0] - u, cen[1]],
        [cen[0] - u/2, cen[1] - 3**(1/2)*u/2],
        [cen[0] + u/2, cen[1] - 3**(1/2)*u/2],
        [cen[0] + u, cen[1]],
        [cen[0] + u/2, cen[1] + 3**(1/2)*u/2],
        [cen[0] - u/2, cen[1] + 3**(1/2)*u/2]
    ]
```

```python
dwg = svgwrite.Drawing('test.svg', profile='full')

basic_u = 25

columns = 15

for i in range(434):
    add_fact = 0
    
    if (i//columns % 2) == 1:
        add_fact = 1.5
        
    text = str(i)
    d_names = {
        395:"CHIL",
        379:"BOLI",
        365:"PARA",
        380:"ARGE",
        349:"PERU",
        335:"BRAS",
        350:"URUG",
        334:"ECUA",
        319:"COLO",
        305:"VENE",
        290:"GUYA",
        306:"SURI",
        291:"GFRA",
        304:"PANA",
        288:"CRIC",
        258:"NICA",
        228:"HOND",
        213:"SALV",
        214:"BELI",
        198:"GUAT",
        168:"MEXI",
        154:"EEUU",
        124:"CANA",
        155:"BAHA",
        185:"CUBA",
        200:"HAIT",
        230:"JAMA",
        216:"RDOM",
        231:"PRIC",
        217:"BARB",
        247:"TYTO"
    }
    
    fill='gray'
    
    if i in d_names:
        
        text = d_names[i]
        #fill = 'purple'
        if text in inv_n_names:
            for range_color in ranges_dict:
                if df.loc[inv_n_names[text]]["value"] >= range_color[0] and df.loc[inv_n_names[text]]["value"] <= range_color[1]:
                    color = ranges_dict[range_color]
            fill = '#%02x%02x%02x' % color
    
        dwg.add(svgwrite.shapes.Polygon(
            polygon_from_centroid(
                    [
                        (i%columns)*3*basic_u + add_fact*basic_u,
                        basic_u*(3**(1/2)/2)*(i//columns)
                    ]
                ,u=basic_u)
                ,fill=fill
                ,stroke='black'
            )
        )

        dwg.add(dwg.text(text, 
                         insert=(
                            (i%columns)*3*basic_u + add_fact*basic_u - basic_u*0.68,
                            basic_u*(3**(1/2)/2)*(i//columns) + basic_u*0.15
                         ),
                         style="font-size:10px"
                        )
               )
f = 0
g = -0
for i, c in enumerate(ranges_dict):
    dwg.add(svgwrite.shapes.Polygon([[620 + g,275-i*25 + f], [695 + g,275-i*25 + f], [695 + g,300-i*25 + f], [620 + g,300-i*25 + f]]
                    ,fill='#%02x%02x%02x' % ranges_dict[c]
                    ,stroke='black'
                )
            )
    dwg.add(dwg.text("%s %%  -  %s %%" % (c[0],
                                    c[1]), 
                         insert=(700+g,293-i*25 + f),
                         style="font-size:12px"
                        )
               )
    
#dwg.add(dwg.text("[]", 
#                         insert=(730,293+25),
#                         style="font-size:12px"
#                        )
#               )

dwg.add(dwg.text("InversiÃ³n en educaciÃ³n como porcentaje de PIB", 
                         insert=(300,130 + f),
                         style="font-size:16px"
                        )
               )

dwg.save()
```
