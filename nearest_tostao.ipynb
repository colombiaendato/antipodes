```python
import folium
import folium.plugins
import pandas as pd
import geopandas
from folium.features import *
from math import sin, cos, sqrt, atan2, radians
from shapely.geometry import asShape, Point, JOIN_STYLE
```

```python
def get_distance(p0, p1):
    R = 6373.0

    lat1 = radians(p0[1])
    lon1 = radians(p0[0])
    lat2 = radians(p1[1])
    lon2 = radians(p1[0])

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    
    return distance
```

```python
df = pd.read_csv("./data/tostaos.csv")
df
```

```python
df['coords'] = df.apply(lambda x:Point(x['lon'], x['lat']),axis=1)
df = geopandas.GeoDataFrame(df, geometry='coords')
```

```python
def nearest(p):
    nearest_index = df["coords"].distance(p).sort_values().index[1]
    nearest_index_point = df.at[nearest_index, 'coords']
    d = get_distance((p.centroid.x, p.centroid.y), (nearest_index_point.centroid.x, nearest_index_point.centroid.y))
    return nearest_index_point, int(d*1000)
```

```python
df['nearest_point'] = df['coords'].apply(nearest).apply(lambda x:x[0])
```

```python
df['nearest_distance'] = df['coords'].apply(nearest).apply(lambda x:x[1])
```

```python
df_nearest = df[df['nearest_distance'] != 0].sort_values('nearest_distance').iloc[0]
df_furthest = df[df['nearest_distance'] != 0].sort_values('nearest_distance', ascending=False).iloc[0]
```

```python
class DivIcon(MacroElement):
    def __init__(self, html='', size=(30,30), anchor=(0,0), style=''):
        """TODO : docstring here"""
        super(DivIcon, self).__init__()
        self._name = 'DivIcon'
        self.size = size
        self.anchor = anchor
        self.html = html
        self.style = style

        self._template = Template(u"""
            {% macro header(this, kwargs) %}
              <style>
                .{{this.get_name()}} {
                    {{this.style}}
                    }
              </style>
            {% endmacro %}
            {% macro script(this, kwargs) %}
                var {{this.get_name()}} = L.divIcon({
                    className: '{{this.get_name()}}',
                    iconSize: [{{ this.size[0] }},{{ this.size[1] }}],
                    iconAnchor: [{{ this.anchor[0] }},{{ this.anchor[1] }}],
                    html : "{{this.html}}",
                    });
                {{this._parent.get_name()}}.setIcon({{this.get_name()}});
            {% endmacro %}
            """)
```
