```python
%matplotlib inline

import json
import matplotlib.pyplot as plt
from shapely.geometry import asShape
from descartes import PolygonPatch
```

```python
data = json.load(open("./data/colombia_not_s_y_p.geojson"))
```

```python
fig = plt.figure(figsize=(8, 10))
ax = fig.gca(xlabel="Longitude", ylabel="Latitude") 

for feat in data["features"]:
    geom = asShape(feat["geometry"])
    x, y = geom.centroid.x, geom.centroid.y
    ax.plot(x, y, '-')
    ax.add_patch(PolygonPatch(feat["geometry"], fc='white', ec='gray',
                alpha=1, lw=1, ls='-', zorder=2))
ax.clear
```
